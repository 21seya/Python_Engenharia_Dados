```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([1,2,3,4,5])
y = x * 2 

plt.xlabel("Vendas")
plt.ylabel("Temperatura")
plt.title("Gráfico de Vendas ",y=0.9,loc="left",x=0.01)

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

font_props = {
    'family':'arial',
    'color':'blue',
    'weight':'bold',
    'size':16
}


x = np.arange(-10,10,0.1)
y = x** 6

plt.text(-5,50,"é uma parábola",fontdict=font_props)

plt.xlabel("Vendas",fontdict=font_props)
plt.ylabel("Temperatura",fontdict=font_props)
plt.title("Vendas",fontdict=font_props)

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

font_props = {
    'family':'arial',
    'color':'blue',
    'weight':'bold',
    'size':16
}

font_props_2 = {
    'family':'arial',
    'color':'green',
    'weight':'normal',
    'size':16
}


x = np.arange(-10,10,0.1)
y = x** 6

plt.text(-5,50,"é uma parábola",fontdict=font_props_2)

plt.xlabel("Vendas",fontdict=font_props)
plt.ylabel("Temperatura",fontdict=font_props)
plt.title("Vendas",fontdict=font_props)

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.arange(0,10,0.5)
y = x** 6 

plt.plot(x,y,c='Red')

plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.title("A função y = x**6")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

x1 = np.arange(0,20,0.5)
y1 = np.cos(x1)

x2 = np.arange(0,20,0.5)
y2 = np.cos(x2) * 2 

plt.xlabel("Vendas")
plt.ylabel("Temperatura")
plt.title("Gráfico de Vendas")

plt.plot(x1,y2,c='g',lw='6.5',ls='dashed',label="Vendas")
plt.plot(x1,y2,c='b',lw='3.5',ls='solid',label="Temperatura")
plt.legend(loc='lower right')
plt.show()

```
