```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([1,2,3,4,5])
y = x * 2 

plt.xlabel("Vendas")
plt.ylabel("Temperatura")
plt.title("Grfico de Vendas ",y=0.9,loc="left",x=0.01)

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

font_props = {
    'family':'arial',
    'color':'blue',
    'weight':'bold',
    'size':16
}


x = np.arange(-10,10,0.1)
y = x** 6

plt.text(-5,50," uma parbola",fontdict=font_props)

plt.xlabel("Vendas",fontdict=font_props)
plt.ylabel("Temperatura",fontdict=font_props)
plt.title("Vendas",fontdict=font_props)

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

font_props = {
    'family':'arial',
    'color':'blue',
    'weight':'bold',
    'size':16
}

font_props_2 = {
    'family':'arial',
    'color':'green',
    'weight':'normal',
    'size':16
}


x = np.arange(-10,10,0.1)
y = x** 6

plt.text(-5,50," uma parbola",fontdict=font_props_2)

plt.xlabel("Vendas",fontdict=font_props)
plt.ylabel("Temperatura",fontdict=font_props)
plt.title("Vendas",fontdict=font_props)

plt.plot(x,y)
plt.show()
```
