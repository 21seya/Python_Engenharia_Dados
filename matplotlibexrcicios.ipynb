```python
import matplotlib.pyplot as plt 
import numpy as np 

font_props = {
    'family':'arial',
    'color':'blue',
    'weight':'bold',
    'size':16
}


x = np.arange(-10,10,0.1)
y = x** 6

plt.text(-5,50," uma parbola",fontdict=font_props)

plt.xlabel("Vendas",fontdict=font_props)
plt.ylabel("Temperatura",fontdict=font_props)
plt.title("Vendas",fontdict=font_props)

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

x1 = np.arange(0,20,0.5)
y1 = np.cos(x1)

x2 = np.arange(0,20,0.5)
y2 = np.cos(x1) + 2 

plt.subplot(1,2,1)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("A primeira funo coseno")
plt.plot(x1,y1,c='g',lw='6.5',ls='dashed',label='y=cos(x)')
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("A segunda funo coseno")
plt.plot(x1,y1,c='b',lw='3.5',ls='solid',label='y=cos(x)*2')
plt.legend(loc='lower right')

plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

valores_esquerda = np.arange(0,6)
valor_direita = np.arange(6,0,-1)
ys = np.arange(6)

plt.barh(ys,valores_esquerda,color='y')
plt.barh(ys,-valor_direita,color='c')

plt.legend(['a','b'])

plt.title("Grfico de barras dividido")
plt.ylabel("Valores de Y")
plt.xlabel("Valores de X")
plt.show()
```
